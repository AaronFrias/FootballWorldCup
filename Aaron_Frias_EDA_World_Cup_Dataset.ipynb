## Creaci贸n de un nuevo DataFrame

A continuaci贸n vamos a trabajar con el DataFrame `df_partidos`. Vamos a centrarnos en la informaci贸n a nivel de selecci贸n, siendo indiferente si el equipo juega en casa o fuera. En este sentido, la informaci贸n disponible no es la mejor para trabajar c贸modamente. Por ello, es habitual realizar transformaciones a los datos en funci贸n de las preguntas que le queramos hacer. A continuaci贸n crearemos un nuevo DataFrame llamado `df_paises`que contendr谩 todos los paises involucrados en la copa del mundo y el a帽o en el que participaron. Es decir, un pa铆s que haya jugado varias copas del mundo aparecer谩 varias veces:


||Year|Country|
|---|---|---|
|0|1930|France|
|1|1930|USA|
|...|...|..|
|37|1938|France|
|...|...|..|


```python
#alternativa 1
df_paises = (
    pd.concat([
        df_partidos[["Year", "Home Team Name"]].rename(columns={"Home Team Name": "Country"}),
        df_partidos[["Year", "Away Team Name"]].rename(columns={"Away Team Name": "Country"})
    ])
)
#alternativa 2
df_paises = (
    df_partidos
    .melt(id_vars=["Year"], value_vars=["Home Team Name", "Away Team Name"])
    .rename(columns={"value": "Country"})
    [["Year", "Country"]]
)
```

```python
df_paises = df_paises.dropna()
df_paises
```

## Calcular el vencedor de cada partido

El DataFrame `df_partidos` tiene la informaci贸n de cada partido, incluyendo los goles marcados por cada equipo, pero no disponemos de una columna que indique si un equipo gan贸 o perdi贸. Vamos a crear una columna llamada `Ganador` que indique el nombre del equipo vencedor (con m谩s goles).

```python
df_partidos["Ganador"] = df_partidos["Home Team Name"]
df_partidos.loc[df_partidos["Home Team Goals"] < df_partidos["Away Team Goals"], "Ganador"] = df_partidos["Away Team Name"]
df_partidos.loc[df_partidos["Home Team Goals"] == df_partidos["Away Team Goals"], "Ganador"] = pd.NA
df_partidos.head()
```

## 11. N煤mero de victorias por pa铆s

Calcularemos para cada pa铆s el n煤mero de partidos ganados.

```python
df_partidos = df_partidos.dropna()
df_partidos.head()
```

```python
(
    df_partidos
    .dropna(subset=["Ganador"])
    .groupby("Ganador")
    .count()
    .sort_values("Year", ascending=False)
)
```

Ahora calcularemos para cada pa铆s el ratio de victorias, calculado como el n煤mero de partidos ganados dividido por el n煤mero de partidos jugados.

```python
df_partidos["MatchID"].value_counts()
```

Podemos observar que tenemos un total de 649 partidos. A continuaci贸n calcularemos el n煤mero de partidos por pa铆s y el n煤mero de paises ganados.

```python
total_partidos = df_paises["Country"].value_counts().sort_values(ascending = False)
print(total_partidos)
```

```python
total_ganados = df_partidos["Ganador"].value_counts().sort_values(ascending = False)
print(total_ganados)
```

```python
# Ordenar por ratio
victorias = (
    df_partidos
    .dropna(subset=["Ganador"])
    .groupby("Ganador")
    .count()
    .rename(columns={"Year":"Victorias"})
    [["Victorias"]]
)

jugados = df_paises.groupby("Country").count().rename(columns={"Year": "Jugados"})[["Jugados"]]
ratio_victorias = pd.merge(jugados, victorias, left_index=True, right_index=True, how="inner")
ratio_victorias["Ratio"] = ratio_victorias["Victorias"] / ratio_victorias["Jugados"]

ratio_victorias.sort_values("Ratio", ascending=False)
```
